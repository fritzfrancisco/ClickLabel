# Visualize ClickLabel Results

```python
import os
import sys

import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
```

```python
# Read Data
data_file = "data/clicks_20250429-214632.csv"
data = pd.read_csv(data_file)
data.head()
```

```python
sns.scatterplot(
    data[(data["Label"] != "Red") & (data["Label"] != "Clutch")],
    x="X",
    y="Y",
    hue="Label",
)
sns.scatterplot(
    data[data["Label"] == "Clutch"],
    x="X",
    y="Y",
    s=800,
    alpha=0.3,
    color="gray",
    label="Clutch",
)
plt.show()
```

```python
df = data[data.Label == "Female"][["X","Y"]]
dm = data[data.Label == "Male"][["X","Y"]]
fig, ax = plt.subplots()
sns.histplot(df, x="X", y="Y", ax=ax, bins=10, color="red")
sns.histplot(dm, x="X", y="Y", ax=ax, bins=10)
plt.show()
```
